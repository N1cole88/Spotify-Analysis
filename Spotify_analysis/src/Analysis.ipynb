```python
from database import Database
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Get the data from the db file
database = Database("first_db", None)
record = database.readTable("track_id", 358)
#print(record)

# Create an empty DataFrame with column names and insert data to df
columns = ['track_id', 'track_name', 'album_name', 'album_popularity', 'release_date', 'artist_name', 'artist_genres', 'acousticness', 'danceability', 'energy', 'instrumentalness', 'liveness', 'loudness', 'mode', 'speechiness', 'tempo', 'valence']
df = pd.DataFrame(columns=columns)

# Iterate over the data and insert into the DataFrame
for row in record:
    df = df.append(pd.Series(row, index=columns), ignore_index=True)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression(fit_intercept=True)
model.fit(X = df[['valence', 'danceability', 'acousticness', 'energy', 'liveness', 'speechiness']], y = df['album_popularity'])
predictions = model.predict(df[['valence', 'danceability', 'acousticness', 'energy', 'liveness', 'speechiness']])
```

```python
df_copy = df
df_copy["predicted_popularity"] = predictions
df_copy["residual"] = df_copy["predicted_popularity"] - df["album_popularity"]
```
